***dropna drops rows/columns containing any missing values.
axis = 0 for rows & axis = 1 for columns***

```python
updated_df = car_df.dropna(axis=0)
```

***Check for data types and inconsistencies***


```python
print(updated_df.info())
```

***Gives summary statistics***

```python
print(updated_df.describe())
```

***This plot helps in understanding the distribution of cars among different makes.***

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=updated_df, x='Make', order=updated_df['Make'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Number of Cars by Make')
plt.xlabel('Make')
plt.ylabel('Count')
plt.show()
```

***This histogram shows the distribution of car prices, which can help identify common price ranges***

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=updated_df, x='Price', bins=20, kde=True, color = 'brown')
plt.title('Distribution of Car Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

***This heatmap displays correlations between numerical variables like Price, Year, Kilometer, etc. It helps identify relationships between variables.***

```python
plt.figure(figsize=(12, 8))
sns.heatmap(updated_df.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

***This pie chart visualizes the distribution of different fuel types among cars.***

```python
desired_fuel_types = ['Petrol', 'CNG', 'Diesel']
filtered_df = updated_df[updated_df['Fuel Type'].isin(desired_fuel_types)]

plt.figure(figsize=(8, 8))
filtered_df['Fuel Type'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Fuel Type Distribution (Petrol, CNG, Diesel)')
plt.ylabel('')
plt.show()
```

***This bar plot gives insight into how the average price varies across different years.***

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=updated_df, x='Year', y='Price')
plt.title('Average Price by Year')
plt.xlabel('Year')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.show()
```

***This count plot shows the distribution of cars by color.***

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=updated_df, x='Color', order=updated_df['Color'].value_counts().index)
plt.title('Number of Cars by Color')
plt.xlabel('Color')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

***This plot shows the distribution of seating capacities among the cars.***

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=updated_df, x='Seating Capacity')
plt.title('Seating Capacity Distribution')
plt.xlabel('Seating Capacity')
plt.ylabel('Count')
plt.show()
```

***This plot displays the distribution of different types of sellers for the cars.***

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=updated_df, x='Seller Type')
plt.title('Seller Type Distribution')
plt.xlabel('Seller Type')
plt.ylabel('Count')
plt.show()
```

***These histograms display the distributions of car lengths, widths, and heights separately, helping understand their variability.***

```python
plt.figure(figsize=(12, 8))
plt.subplot(3, 1, 1)
sns.histplot(data=updated_df, x='Length', bins=20, kde=True)
plt.title('Distribution of Length')

plt.subplot(3, 1, 2)
sns.histplot(data=updated_df, x='Width', bins=20, kde=True)
plt.title('Distribution of Width')

plt.subplot(3, 1, 3)
sns.histplot(data=updated_df, x='Height', bins=20, kde=True)
plt.title('Distribution of Height')

plt.tight_layout()
plt.show()
```

***This barplot provides comparison of average prices across different locations.***

```python
plt.figure(figsize=(50, 30))
sns.barplot(data=updated_df, x='Location', y='Price', errorbar=None)
plt.title('Average Price by Location')
plt.xlabel('Location')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.show()
```

***This plot will show the count of cars for each transmission type.***

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=updated_df, x='Transmission')
plt.title('Count of Cars by Transmission Type')
plt.xlabel('Transmission Type')
plt.ylabel('Count')
plt.show()
```
